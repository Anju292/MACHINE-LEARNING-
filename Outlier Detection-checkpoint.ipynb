## 2.IQR METHOD( Interquartile Range (IQR) Method)

This method is useful for data that do not necessarily follow a normal distribution.

Steps:

- Calculate the first (Q1) and third (Q3) quartiles of the data.
- Compute the IQR (IQR = Q3 - Q1).
- Define the lower and upper bounds as Q1 - 1.5IQR and Q3 + 1.5IQR.
- Identify outliers as data points falling outside these bounds.

## 3.Percentile Method for Outlier Detection

The percentile method is a robust technique for handling outliers, especially in datasets that do not follow a normal distribution. Here are the key points:

- **Percentile Selection:** Choose lower and upper percentiles (commonly 1st and 99th, or 5th and 95th percentiles).
- **Determine Thresholds:** Calculate the data values at these selected percentiles.
- **Cap Extreme Values:** Adjust data points below the lower percentile to the value at the lower percentile, and data points above the upper percentile to the value at the upper percentile.
- **Maintain Data Range:** This method retains the bulk of the data while mitigating the impact of extreme values, ensuring the dataset remains within a practical range.

The percentile method is particularly useful for skewed data or when you need to limit the influence of outliers without completely removing data points.

### 4.NORMAL DISTRIBUTION METHOD(Z-Score Method)

Normal Distribution Method and the Z-Score Method are essentially the same when it comes to detecting outliers. Both methods rely on the assumption that the data follows a normal distribution and use the standard deviation as a measure of dispersion. Here are the key points:
- **Assumption:** The data follows a normal distribution.
- **Calculation:** Outliers are identified based on the distance from the mean.
  - **Z-Score Calculation:** \[ \text{z} = \frac{(X - \mu)}{\sigma} \]
    - \( X \) is the data point.
    - \( \mu \) is the mean of the data.
    - \( \sigma \) is the standard deviation of the data.
- **Threshold:** Commonly, data points with z-scores greater than 3 or less than -3 are considered outliers.
- **Interpretation:** A z-score indicates how many standard deviations a data point is from the mean. For example, a z-score of 3 means the data point is 3 standard deviations above the mean.



```python
# Calculate mean, median, and standard deviation

mean_price_per_sqft = df['price_per_sqft'].mean()
median_price_per_sqft = df['price_per_sqft'].median()
std_price_per_sqft = df['price_per_sqft'].std()
```

```python
# Calculate upper and lower limits
upper_limit = mean_price_per_sqft + 3 * std_price_per_sqft
lower_limit = mean_price_per_sqft - 3 * std_price_per_sqft

# Display upper and lower limits
print("Upper Limit:", upper_limit)
print("Lower Limit:", lower_limit)
```

```python
# Identify outliers
outliers = df[(df['price_per_sqft'] > upper_limit) | (df['price_per_sqft'] < lower_limit)]
# Remove outliers
new_df = df[(df['price_per_sqft'] <= upper_limit) & (df['price_per_sqft'] >= lower_limit)]
```
